### Networks between characters from act 1 to act 5

```python
characters = [
'Cap.', 
'Wife.', 
'Jul.', 
'Tyb.', 
'Nurse.', 
'Pet.', 
'Greg.', 
'Samp.', 
'Anthony', 
'Potpan', 
'Serv.',
'Mon.',
'M. Wife.',
'Rom.', 
'Ben.', 
'Bal.', 
'Abr.', 
'Prince.',
'Par.',
'Mer.', 
'Laur.', 
'John.', 
'Chor.', 
'Apoth.', 
'Chief Watch.', 
'Musicians',
'Verona.'
]
```

```python
character_numbers = {}

connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    character_numbers[k] = 0

for i in acts:
    for j in i.split('Scene')[1:]:
        scenecharacters = []
        for k in characters:
            if j.find(k) != -1:
                scenecharacters.append(k)
                character_numbers[k] += 1
        for a in range(len(scenecharacters)-1):
            for b in range(a+1,len(scenecharacters)):
                connections[(scenecharacters[a],scenecharacters[b])] += 1
```

```python
character_numbers
```

Actors and the number of the scenes that they occure.

```python
connections
```

The number of scenes that characters appear together. 

```python
for key,value in character_numbers.items():
    print(key,value)
```

```python
graphitems = []

for key,value in character_numbers.items():
    dashnode = {'data': {'id': key,
                         'label': key.title(),
                         'size': str(value)}}
    graphitems.append(dashnode)

for key,value in connections.items():
    if value != 0:
        dashedge = {'data': {'source': key[0],
                             'target': key[1],
                             'weight': value}}
        graphitems.append(dashedge)
    
graphitems
```

```python
graphitems = []

for key,value in character_numbers.items():
    dashnode = {'data': {'id': key,
                         'label': key.title(),
                         'size': str(value)}}
    graphitems.append(dashnode)

correction_factor = max(connections.values())
print(correction_factor)
for key,value in connections.items():
    if value != 0:
        dashedge = {'data': {'source': key[0],
                             'target': key[1],
                             'weight': value/correction_factor}}
        graphitems.append(dashedge)
    
graphitems
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    cyto.Cytoscape(
        layout={'name': 'cose'},
        elements=graphitems,
        style={'width': '100%', 'height': '750px'},
        stylesheet=[
            {
                'selector': 'node',
                'style': {
                    'content':'data(label)',
                    'text-halign':'center',
                    'text-valign':'center',
                    'width': 'data(size)',
                    'height': 'data(size)',
                    'font-size':4,
                    'color': 'black',
                    'background-color': 'green',
                    'text-outline-color': 'white',
                    'text-outline-width': 0.2,
                    'shape':'square'
                }
            },
            {
                'selector':'edge',
                'style': {
                    'width':'data(weight)',
                    'line-color': 'black',
                  }
            },
        ]
    )
])

app.run_server(debug=True)
```

According to this graph, Romeo, Juliet, Juliet's nurse, Capulet (Juliet's father), Benvolio (Romeo's coursin), and Paris (whom Capulet hopes Juliet marries) are most important figures in this play. But this result is only build upon the number of these figures' occurrance. If there's a character who only appears few times but plays an important role in the play, this graph would completely ignore that figure.

```python
from networkx.algorithms.community.centrality import girvan_newman
```

```python
communities = girvan_newman(G)
```

```python
node_groups = []
for com in next(communities):
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
nx.density(G)
```

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
greedy_modularity_communities(G)
```

```python
communities = greedy_modularity_communities(G)

modularity_class = {}

for community_number, community in enumerate(communities):
    for name in community:
        modularity_class[name] = community_number
```

```python
modularity_class
```

```python
communities = greedy_modularity_communities(G)

node_groups = []
for com in communities:
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    elif node in node_groups[1]:
        color_map.append('yellow')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
graphitems = []

for k,v in character_numbers.items():
    if k in node_groups[0]:
        modularity = 'blue'
    elif k in node_groups[1]:
        modularity = 'green'
    else:
        modularity = 'red'
    dashnode = {'data': {'id': k,
                         'label': k.title(),
                         'size': str(v),'modularity': modularity}}
    graphitems.append(dashnode)

correction_factor = max(connections.values())
print(correction_factor)
for k,v in connections.items():
    if v != 0:
        dashedge = {'data': {'source': k[0],
                             'target': k[1],
                             'weight': v/correction_factor}}
        graphitems.append(dashedge)
    
graphitems
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    cyto.Cytoscape(
        layout={'name': 'cose'},
        elements=graphitems,
        style={'width': '100%', 'height': '750px'},
        stylesheet=[
            {
                'selector': 'node',
                'style': {
                    'content':'data(label)',
                    'text-halign':'center',
                    'text-valign':'center',
                    'width': 'data(size)',
                    'height': 'data(size)',
                    'font-size': 4,
                    'color': 'black',
                    'background-color': 'data(modularity)',
                    'text-outline-color': 'white',
                    'text-outline-width': 0.2,
                    'shape':'circle'
                }
            },
            {
                'selector':'edge',
                'style': {
                    'width':'data(weight)',
                    'line-color': 'blue',
                  }
            },
        ]
    )
])

app.run_server(debug=True)
```
