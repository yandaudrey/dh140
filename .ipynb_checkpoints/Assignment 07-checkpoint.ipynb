```python
import requests
from bs4 import BeautifulSoup
```

```python
RJ = requests.get('https://www.gutenberg.org/files/1513/1513-h/1513-h.htm')
```

```python
RJ
```

```python
html_string = RJ.text
document = BeautifulSoup(html_string, "html.parser")
```

```python
document
```

```python
print(document.text)
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent_RJ = sent_tokenize(document.text) 
```

```python
sent_RJ[0]
```

```python
print(word_tokenize(sent_RJ[0]))
```

```python
words = []
for s in sent_RJ:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation 
nltk.download('stopwords')
```

```python
extra_stopwords = ['I',"'s",'And',"'d",'Gutenberg','eBook','United', 'States']
```

```python
mystopwords = list(punctuation) + stopwords.words('english') + extra_stopwords
```

```python
RJ_Nostopwords = []
for i in words:
    if i not in mystopwords: 
        RJ_Nostopwords.append(i)
print(RJ_Nostopwords)
```

### Word frequencies and the mean, median, mode, and trimmed mean

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(RJ_Nostopwords)
```

```python
freq
```

```python
import pandas as pd
```

```python
type(freq)
```

```python
data_items = freq.items()
data_list = list(data_items)
df = pd.DataFrame(data_list,columns=['word','freq'])
print(df)
```

```python
df.describe()
```

- mean = 4.122116
- median = 1.000000
- mode = 873.000000

```python
df2 = df.sort_values(by='freq',ignore_index=True).copy()
```

```python
df2
```

```python
df2.loc[int(0.2*4593):int(0.8*4593)].mean()
```

- trimmed mean(remove the smalles 20% and the largest 20%): 1.503446

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = featrues_df[['s5']]
y = target_df
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```

```python
X = featrues_df[['age']]
y = target_df
```

```python
reg_new = LinearRegression().fit(X, y)
```

```python
ytrain = reg_new.intercept_ + reg_new.coef_ * X
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```
